### Repeat without book

```python
weight = 0.5
input = 2
alpha = 0.1
goal_pred = 0.8

for i in range(100):
    pred = weight * input
    error = (pred - goal_pred)**2
    derivative = input * (pred - goal_pred)
    weight = weight - (derivative * alpha)

    print("Error = " + str(error) + " Weight = " + str(weight) + " Prediction = " + str(pred) + "\n")
```
