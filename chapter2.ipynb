### Error measurement

```python
knob_weight = 0.5
input = 0.5
goal_pred = 0.8

pred = knob_weight * input

error = (pred - goal_pred) ** 2
print(error)
```

### Heat/cold

```python
weight = 0.5
input = 0.5
goal_pred = 0.8

step_amount = 0.001

for iteration in range(2000):
    print("Iteration: " + str(iteration) + " Weight:" + str(weight))
    pred = weight * input
    error = (pred - goal_pred) ** 2

    print("Error:" + str(error) + " Pred:" + str(pred))

    up_prediction = input * (weight + step_amount)
    up_error = (goal_pred - up_prediction) ** 2

    down_prediction = input * (weight - step_amount)
    down_error = (goal_pred - down_prediction) ** 2

    print("Up:" + str(up_prediction) + " Up err:" + str(up_error))
    print("Down:" + str(down_prediction) + " Down err:" + str(down_error))
    print("Is errors are equal?" + str(down_error == up_error) + "\n")
    if down_error < up_error:
        weight = weight - step_amount

    if down_error > up_error:
        weight = weight + step_amount
```

### Defining direction of weight change based on error value(gradient)

```python
weight = 0.5
goal_pred = 0.8
input = 0.5

for iteration in range(20):
    pred = weight * input
    error = (pred - goal_pred) ** 2
    direction_and_amount = (pred - goal_pred) * input
    weight = weight - direction_and_amount

    print("Iteration: " + str(iteration) + "\nError:" + str(error) + "\nPred:" + str(pred))
```

```python
weight, goal_pred, input = 0.0, 0.8, 1.1

for iteration in range(4):
    pred = weight * input
    error = (pred - goal_pred) ** 2
    delta = pred - goal_pred
    weight_delta = delta * input
    weight = weight - weight_delta
    print("Error:" + str(error) + "\nPred:" + str(pred) + "\nWeight delta:" + str(weight_delta))
```

### Alpha coefficient

```python
weight = 0.5
goal_pred = 0.8
input = 10
alpha = 0.01

for iteration in range(100):
    pred = weight * input
    error = (pred - goal_pred) ** 2
    derivative = (pred - goal_pred) * input
    weight = weight - (derivative * alpha)

    print("Error: " + str(error) + "\nPred:" + str(pred) + "\n")
```
